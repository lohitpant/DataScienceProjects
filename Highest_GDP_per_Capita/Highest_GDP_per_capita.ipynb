# Find and compare countries with Highest GDP per capita

Step 1: Importing necessary libraries and data set in .csv format

```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('nations.csv')
data.head()
```

Step 2: Filling N/A values with median values for 'gdp_percp' column

```python
data['gdp_percap'].fillna(data['gdp_percap'].median(), inplace=True)
data.head(10)
```

Step 3: Grouping the data by 'country' column and finding mean of it for each country

```python
avggdp_percap = data.groupby(['country']).mean()['gdp_percap']
avggdp_percap
```

Step 4: Sorting the data in Decreasing order as per 'gdp_percap' and printing top 5 countries

```python
top_five=avggdp_percap.sort_values(ascending=False).head(5)
top_five
```

Step 5: We will save 'United Arab Emirates' country data in 'uae' variable aftr sorting it 'year' wise

```python
uae=data[data['country']=='United Arab Emirates'].sort_values('year')
uae
```

Step 6: Now we will plot 'uae' data with 'year' on x axis and 'gdp_percap' on y axis.

```python
axis= plt.plot(uae['year'], uae['gdp_percap'])
axis=set_xlabel='Year'
axis=set_ylabel='GDP Per Capita'
```
